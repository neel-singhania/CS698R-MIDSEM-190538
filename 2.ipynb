# testing the policy evaluation

```python
def policyEvaluation(pi,P,gamma,theta):
    vold = np.zeros((3,4))
    while 1:
        vnew = np.zeros((3,4))
        for i in range(3):
            for j in range(4):
                for a in range(4):
                    temp = np.zeros((3,4))
                    for k in range(len(P[i][j][a])):
                        p = P[i][j][a][k][0]
                        s = P[i][j][a][k][1]
                        r = P[i][j][a][k][2]
                        f = P[i][j][a][k][3]
#                         print(p,s,r,f)
                        
                        temp[i][j] += p*(r+gamma*vold[s//4][s%4])
                    if a == pi[i][j]:
                        vnew += temp
                    else:
                        vnew += 0
        if(np.amax(vnew-vold))<theta:
            break
        vold = vnew
    return vnew
```

```python
print(policyEvaluation(pi,P,gamma,theta))
```

# policy improvement and testing it

```python
def policyImprovement(v,P,gamma):
    Q = np.zeros((3,4,4))
    pi = np.zeros((3,4))
    for i in range(3):
        for j in range(4):
            for a in range(4):
                for k in range(len(P[i][j][a])):
                    p = P[i][j][a][k][0]
                    s = P[i][j][a][k][1]
                    r = P[i][j][a][k][2]
                    f = P[i][j][a][k][3]
#                         print(p,s,r,f)
                    
                    Q[i][j][a] += p*(r+gamma*v[s//4][s%4])
        
    for i in range(3):
        for j in range(4):
            pi[i][j] = np.argmax(Q[i][j])
    return pi
```

```python
print(policyImprovement(policyEvaluation(pi,P,gamma,theta),P,gamma))
```

# policy iteration

```python
def policyIteration(pi,P,gamma,theta):
#     pi = random.randint(0,4,size = (3,4))
    cnt = 0
    while 1:
        cnt+=1
        piold = pi
        v = policyEvaluation(pi,P,gamma,theta)
        pi = policyImprovement(v,P,gamma)
        if np.all(piold==pi):
            break
            
    return v,pi,cnt
```

```python
policyIteration(pi,P,gamma,theta)
```

# Value Iteration

```python
def valueIteration(P,gamma,theta):
    cnt = 0
    v = np.zeros((3,4))
    while 1:
        cnt += 1
        Q = np.zeros((3,4,4))
        for i in range(3):
            for j in range(4):
                for a in range(4):
                    for k in range(len(P[i][j][a])):
                        p = P[i][j][a][k][0]
                        s = P[i][j][a][k][1]
                        r = P[i][j][a][k][2]
                        f = P[i][j][a][k][3]
                        
                        Q[i][j][a] += p*(r+gamma*v[s//4][s%4])
        maxi = 0
        for i in range(3):
            for  j in range(4):
                maxi = max(maxi,abs(v[i][j]-np.amax(Q[i][j])))
        if maxi<theta:
            break
        
        for i in range(3):
            for j in range(4):
                v[i][j] = np.amax(Q[i][j])
                pi[i][j] = np.argmax(Q[i][j])
    return v,pi,cnt
```
